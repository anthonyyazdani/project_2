```python
import os
from functions.classical_ML_functions import *

ORIGINAL_DIR = os.path.dirname(os.getcwd())
DIR = ORIGINAL_DIR + "\\project_2\\embeddings\\glove\\stanford_glove"
DATA_DIR = ORIGINAL_DIR + "\\project_2"
SAVING_DIR = ORIGINAL_DIR + "\\project_2\\Classical_ML_results\\glove_twitter_200d"
```

# Glove twitter 200d

```python
embeddings_index = create_dict_from_glove(DIR + '\\glove.twitter.27B.200d.txt')
```

# Self made Data

```python
# Importing data

data, labels, test = load_data(DATA_DIR + '\\self_made_data\\self_made_positive.txt',
                               DATA_DIR + '\\self_made_data\\self_made_negative.txt',
                               DATA_DIR + '\\self_made_data\\self_made_test.txt')
# Convert to tokens

data = tokenize(data)
test = tokenize(test)

# Take the average vectors of words to create the feature matrix

data_features = dict_embed(data, embeddings_index, 200)
test_features = dict_embed(test, embeddings_index, 200)
```

###### Naive bayes

```python
# KFold

self_made_NB_results = stratified_kfold_NB(data_features, labels)
msmnb = np.mean(self_made_NB_results)
stdsmnb = np.std(self_made_NB_results)
print(f"Mean accuracy = {round(msmnb*100, 4)}%, Std accuracy = {round(stdsmnb, 4)}")
```

```python
# Prediction

clf = GaussianNB().fit(data_features, labels)
predsmnb = clf.predict(test_features)
predsmnb[np.where(predsmnb==0)] = -1
```

###### Logistic regression

```python
# KFold

self_made_LR_results = stratified_kfold_LR(data_features, labels)
msmlr = np.mean(self_made_LR_results)
stdsmlr = np.std(self_made_LR_results)
print(f"Mean accuracy = {round(msmlr*100, 4)}%, Std accuracy = {round(stdsmlr, 4)}")
```

```python
# Prediction

clf = LogisticRegression(max_iter=1000).fit(data_features, labels)
predsmlr = clf.predict(test_features)
predsmlr[np.where(predsmlr==0)] = -1
```

###### Multi layer perceptron

```python
# KFold

self_made_MLP_results = stratified_kfold_MLP(data_features, labels)
msmmlp = np.mean(self_made_MLP_results)
stdsmmlp = np.std(self_made_MLP_results)
print(f"Mean accuracy = {round(msmmlp*100, 4)}%, Std accuracy = {round(stdsmmlp, 4)}")
```

```python
# Prediction

MLPClassifier(hidden_layer_sizes=(100, 100),
              random_state=1,
              max_iter=1000,
              verbose=False,
              early_stopping=True).fit(data_features, labels)
predsmmlp = clf.predict(test_features)
predsmmlp[np.where(predsmmlp==0)] = -1
```
